```
import json
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
def GetPolygon(loc):
	with open(loc, "r") as f:
		data=json.load(f)
	f.close()

	data = data['objects']
	data = [x['polygon'] for x in data]
	data = [np.int32(x) for x in data]
	return data
```

```
def GetSegments(loc, data):
  img = cv2.imread(loc, 1)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  for x in data:
    cv2.polylines(img, [x], True, (0,255,155), 3)
  return img
```
